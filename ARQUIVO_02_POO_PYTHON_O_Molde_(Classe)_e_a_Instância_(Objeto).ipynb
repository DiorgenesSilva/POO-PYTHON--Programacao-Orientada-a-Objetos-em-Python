### **Módulo 2: O Molde (Classe) e a Instância (Objeto)**

image.png

```python
class Personagem:
    # O "Construtor" - Define o que o objeto TEM ao nascer
    def __init__(self, nome, classe_rpg, nivel):
        self.nome = nome          # Atributo
        self.classe_rpg = classe_rpg  # Atributo
        self.nivel = nivel        # Atributo

    # Um Método - Define o que o objeto FAZ
    def saudar(self):
        print(f"Olá! Eu sou {self.nome}, um {self.classe_rpg} de nível {self.nivel}.")
```

image.png

```python
# Criando os objetos (instâncias)
heroi_01 = Personagem("Aragon", "Guerreiro", 10)
heroi_02 = Personagem("Gandalf", "Mago", 20)

# Usando os objetos
print(heroi_01.saudar())
print(heroi_02.saudar())
heroi_01.saudar()
heroi_02.saudar()
```
